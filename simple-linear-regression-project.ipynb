```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
```

```python
dataset = pd.read_csv('/kaggle/input/salarydata/Salary_Data.csv')
```

```python
X = dataset.iloc[:, :-1].values   # YearsExperience
Y = dataset.iloc[:, 1].values     # Salary
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=0)
```

```python
regressor = LinearRegression()       # Create model (instance)
regressor.fit(X_train, Y_train)      # Train model on training set
```

```python
Y_pred = regressor.predict(X_test)
```

```python
plt.figure(figsize=(8,5))
plt.scatter(X_train, Y_train, color='green', label='Actual Salary')
plt.plot(X_train, regressor.predict(X_train), color='#39ff14', label='Predicted Regression Line')
plt.title('Salary vs Experience (Training Set)')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.scatter(X_test, Y_test, color='red', label='Actual Test Salary')
plt.plot(X_train, regressor.predict(X_train), color='blue', label='Predicted Regression Line')
plt.title('Salary vs Experience (Test Set)')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.legend()
plt.show()
```

```python
r2 = r2_score(Y_test, Y_pred)
print(f"✅ Model R² Score: {r2:.4f}")
```
